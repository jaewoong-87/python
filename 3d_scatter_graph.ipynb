```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

엑셀의 데이터 값들 가져오기

```python
x_axis =pd.read_excel('3d_data_graph.xlsx', usecols=[0])
y_axis =pd.read_excel('3d_data_graph.xlsx', usecols=[2])
z_axis =pd.read_excel('3d_data_graph.xlsx', usecols=[1])

x_axis1 =pd.read_excel('stress.xlsx', usecols=[0])
y_axis1 =pd.read_excel('stress.xlsx', usecols=[2])
z_axis1 =pd.read_excel('stress.xlsx', usecols=[1])
```

```python
fig = plt.figure(figsize=(10,10))

ax = fig.gca(projection='3d')

ax.scatter(x_axis,y_axis,z_axis, marker='o', s=15, c='darkgreen')

plt.xlabel('concentration')
plt.ylabel('shear_rate')

plt.title("The viscosity depending on concentration and shear_rate")
plt.show()
```

```python
fig1 = plt.figure(figsize=(10,10))

ax = fig1.gca(projection='3d')

ax.scatter(x_axis1,y_axis1,z_axis1, marker='x', s=15, c='darkblue')

plt.xlabel('concentration')
plt.ylabel('shear_rate')

plt.title("The stress depending on concentration and shear_rate")
plt.show()
```
